```python
import matplotlib.pyplot as plt

train_errors = []
val_losses = []

with open('log/log_mamba.txt', 'r') as f:
    for line in f:
        parts = line.split()
        if len(parts) == 3:  
            step, mode, error = int(parts[0]), parts[1], float(parts[2])
            if mode == 'train':
                train_errors.append(error)
            elif mode == 'val':
                val_losses.append((step, error))

# Sort the validation losses by step
val_losses.sort(key=lambda x: x[0])

# Plot training and validation errors
plt.figure(figsize=(8, 6))
plt.plot(range(len(train_errors)), train_errors, label='Training Error')
x = [x[0] for x in val_losses]
y = [x[1] for x in val_losses]
plt.plot(x, y, label='Validation Loss', color='red', marker='x')

plt.xlabel('Step (#)')
plt.ylabel('Error/Loss')
plt.title('Training and Validation Errors/Losses')
plt.legend()
plt.show()
```
